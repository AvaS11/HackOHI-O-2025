```python
# Merging Data from lines and flows

# Pandas
import pandas as pd

# Read CSV files
lines = pd.read_csv('lines.csv')
flows = pd.read_csv('line_flows_nominal.csv')

# Merge DataFrames on 'name' column
df = lines.merge(flows, on='name')
df['loading_percent'] = df['p0_nominal'] / df['s_nom'] * 100

# Display the first few rows of the merged DataFrame
df.head()
```

```python
#EDA Histograms (numbers)

#specify number types
num_columns = df.select_dtypes(include='number').columns

#make histograms (pandas will automatically size the grid)
df[num_columns].hist(bins=15, figsize=(15,6))

#alter layout
plt.tight_layout()
plt.show()
```

```python
#EDA Histograms (objects)

#specify object data type
categorical_columns = df.select_dtypes(include='object').columns

#iterate through columns
for col in categorical_columns:
    plt.figure(figsize=(10,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()
```
